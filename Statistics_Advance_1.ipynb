# *Statistics Advance - Part 1*

# 1. What is a random variable in probability theory?

A random variable is a numerical outcome of a random process or experiment.
# 2. What are the types of random variables?
There are two types: Discrete (countable outcomes) and Continuous (measurable outcomes).
#3. What is the difference between discrete and continuous distributions?
Discrete distributions involve countable outcomes, while continuous distributions involve outcomes over an interval.
#4. What are probability distribution functions (PDF)?
PDFs show the probability of a random variable taking a specific value (discrete) or falling within a range (continuous).
#5. How do cumulative distribution functions (CDF) differ from probability distribution functions (PDF)?
CDF gives the probability that a variable is less than or equal to a value. PDF shows probability for a specific value or range.
#6. What is a discrete uniform distribution?
It assigns equal probabilities to all outcomes in a finite set, like rolling a fair die.
#7. What are the key properties of a Bernoulli distribution?
Only two outcomes (success/failure), probability of success = p, failure = 1 - p.
#8. What is the binomial distribution, and how is it used in probability?
It models the number of successes in fixed independent trials with the same success probability.
#9. What is the Poisson distribution and where is it applied?
It models the number of events in a fixed interval of time or space when events occur independently. Example: calls per hour.
#10. What is a continuous uniform distribution?
A distribution where all values in a range are equally likely. Example: selecting a random number between 0 and 1.
#11. What are the characteristics of a normal distribution?
Bell-shaped, symmetric around the mean, mean = median = mode, defined by mean and standard deviation.
#12. What is the standard normal distribution, and why is it important?
A normal distribution with mean 0 and standard deviation 1. It helps compare different datasets using Z-scores.
#13. What is the Central Limit Theorem (CLT), and why is it critical in statistics?
CLT states that the sampling distribution of the sample mean approaches a normal distribution as sample size grows, regardless of the population’s distribution.
#14. How does the Central Limit Theorem relate to the normal distribution?
It justifies using the normal distribution for inference even when the population isn't normally distributed.
#15. What is the application of Z statistics in hypothesis testing?
Z statistics help determine if a sample mean significantly differs from a population mean.
#16. How do you calculate a Z-score, and what does it represent?
Z = (X - μ) / σ. It represents how many standard deviations a value is from the mean.
#17. What are point estimates and interval estimates in statistics?
Point estimate gives a single value (e.g., sample mean). Interval estimate gives a range (e.g., confidence interval).
#18. What is the significance of confidence intervals in statistical analysis?
They indicate the range in which a population parameter likely falls with a certain confidence level (e.g., 95%).
#19. What is the relationship between a Z-score and a confidence interval?
Z-scores help determine the boundaries of confidence intervals for normal distributions.
#20. How are Z-scores used to compare different distributions?
They standardize values, allowing comparisons across different scales or distributions.
#21. What are the assumptions for applying the Central Limit Theorem?
Independent samples, sufficiently large sample size (typically n > 30), finite variance.
#22. What is the concept of expected value in a probability distribution?
It’s the long-run average value of repetitions of the experiment, calculated as Σ[x * P(x)].
#23. How does a probability distribution relate to the expected outcome of a random variable?
The probability distribution provides the weights (probabilities) used to compute the expected outcome.


# ***Practical***



```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats

# Set plot style
sns.set(style="whitegrid")


```

# 1. Generate random variable and display value

```python
random_value = np.random.rand()
print("1. Random Value:", random_value)
```

# 2. Discrete uniform distribution & PMF

```python
discrete_data = np.random.randint(1, 7, 1000)
values, counts = np.unique(discrete_data, return_counts=True)
pmf = counts / counts.sum()
plt.figure(figsize=(6, 4))
plt.bar(values, pmf)
plt.title("2. PMF of Discrete Uniform Distribution")
plt.xlabel("Value")
plt.ylabel("Probability")
plt.show()
```

# 3. Bernoulli PDF function

```python
def bernoulli_pdf(p, x):
    return p if x == 1 else 1 - p
```

# 4. Simulate binomial distribution

```python
binom_data = np.random.binomial(n=10, p=0.5, size=1000)
plt.hist(binom_data, bins=11, edgecolor='black')
plt.title("4. Binomial Distribution Histogram (n=10, p=0.5)")
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.show()
```

# 5. Poisson distribution

```python
poisson_data = np.random.poisson(lam=3, size=1000)
sns.histplot(poisson_data, kde=False, bins=range(10), edgecolor='black')
plt.title("5. Poisson Distribution")
plt.show()
```

# 6. CDF of discrete uniform

```python
cdf = np.cumsum(pmf)
plt.step(values, cdf, where='mid')
plt.title("6. CDF of Discrete Uniform Distribution")
plt.xlabel("Value")
plt.ylabel("CDF")
plt.show()
```

# 7. Continuous uniform distribution

```python
cont_uniform = np.random.uniform(0, 1, 1000)
sns.histplot(cont_uniform, kde=True, bins=20)
plt.title("7. Continuous Uniform Distribution")
plt.show()
```

# 8. Normal distribution histogram

```python
normal_data = np.random.normal(loc=0, scale=1, size=1000)
sns.histplot(normal_data, kde=True)
plt.title("8. Normal Distribution Histogram")
plt.show()
```

# 9. Z-scores and plot

```python
z_scores = stats.zscore(normal_data)
plt.plot(z_scores[:100])
plt.title("9. Z-Scores Plot")
plt.xlabel("Index")
plt.ylabel("Z-Score")
plt.show()
```

# 10. CLT with exponential distribution

```python

means = [np.mean(np.random.exponential(scale=2, size=30)) for _ in range(1000)]
sns.histplot(means, kde=True)
plt.title("10. CLT with Exponential Distribution")
plt.show()
```

# 15. Multiple samples from normal for CLT

```python
sample_means = [np.mean(np.random.normal(0, 1, 30)) for _ in range(1000)]
sns.histplot(sample_means, kde=True)
plt.title("15. CLT with Normal Samples")
plt.show()
```


# 16. Standard normal PDF plot

```python
x = np.linspace(-4, 4, 1000)
pdf = stats.norm.pdf(x, 0, 1)
plt.plot(x, pdf)
plt.title("16. Standard Normal Distribution")
plt.xlabel("Z")
plt.ylabel("PDF")
plt.show()
```

# 17. Binomial probabilities

```python

binom_probs = stats.binom.pmf(range(11), 10, 0.5)
plt.bar(range(11), binom_probs)
plt.title("17. Binomial PMF")
plt.show()
```


# 18. Z-score for data point

```python
value = 75
mean = 70
std = 5
z_score = (value - mean) / std
print("18. Z-score of 75:", z_score)
```

# 19. Hypothesis test (Z-test)

```python

sample = np.random.normal(52, 5, 30)
z_stat, p_value = stats.ttest_1samp(sample, popmean=50)
print("19. Hypothesis Test: Z-stat =", z_stat, ", p-value =", p_value)
```


# 20. Confidence interval

```python
conf_int = stats.norm.interval(0.95, loc=np.mean(sample), scale=stats.sem(sample))
print("20. 95% Confidence Interval:", conf_int)
```



# 21. Normal data + CI for mean

```python
norm_sample = np.random.normal(100, 10, 50)
ci = stats.norm.interval(0.95, loc=np.mean(norm_sample), scale=stats.sem(norm_sample))
print("21. CI for normal sample mean:", ci)
```


# 22. Normal PDF

```python
x = np.linspace(60, 140, 1000)
pdf = stats.norm.pdf(x, loc=100, scale=10)
plt.plot(x, pdf)
plt.title("22. Normal PDF")
plt.show()
```

# 23. Poisson CDF

```python
x = np.arange(0, 15)
cdf = stats.poisson.cdf(x, mu=3)
plt.step(x, cdf, where='mid')
plt.title("23. Poisson CDF")
plt.show()
```

# 24. Continuous uniform expected value

```python

uniform_data = np.random.uniform(10, 20, 1000)
expected_value = np.mean(uniform_data)
print("24. Expected Value of Uniform(10,20):", expected_value)
```

# 25. Compare std of two datasets

```python
data1 = np.random.normal(0, 1, 100)
data2 = np.random.normal(0, 5, 100)
plt.boxplot([data1, data2], labels=["Std=1", "Std=5"])
plt.title("25. Compare Standard Deviations")
plt.show()
```


# 26. Range and IQR

```python
data = np.random.normal(0, 1, 100)
data_range = np.ptp(data)
iqr = stats.iqr(data)
print("26. Range:", data_range, ", IQR:", iqr)
```


# 27. Z-score normalization

```python

z_normalized = stats.zscore(data)
plt.hist(z_normalized, bins=20, edgecolor='black')
plt.title("27. Z-Score Normalization")
plt.show()
```

# 28. Skewness and Kurtosis

```python
skew = stats.skew(data)
kurt = stats.kurtosis(data)
print("28. Skewness:", skew, ", Kurtosis:", kurt)
```
